```python
import pandas as pd
import pyarrow as pa
import pyarrow.parquet as pq
```

```python
df = pd.read_parquet('lyrics_data.parquet')
```

```python
table = pa.Table.from_pandas(df)
```

```python
df = df[df['language']=='en']
```

```python
df
```

```python
df.reset_index(inplace=True)
```

```python
df.drop('index', axis=1, inplace=True)
```

```python
df['ALink'] = df['ALink'].str.replace('/', '')
df['Lyric'] = df['Lyric'].str.replace('\n', ' ')
```

```python
df.drop('SLink', axis=1, inplace=True)
```

```python
df.to_parquet('lyrics.parquet', compression='brotli')
```
