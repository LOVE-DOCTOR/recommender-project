```python
# Set the API key and base URL for the TMDb API
api_key = 'de0ce2956064588ee289cb8da1cc9a77'
base_url = 'https://api.themoviedb.org/3/trending/all/day'

# Initialize an empty DataFrame to store the movie data
movies_df = pd.DataFrame()

# Set the maximum number of pages to retrieve
max_pages = 1005  # 20 movies per page
max_pages = int(max_pages)

# Set the language parameter for the API request
language = 'en-US'

# Loop through the pages and retrieve the movie data
for page in range(1, max_pages+1):
    # Construct the API request URL
    url = f'{base_url}?api_key={api_key}&language={language}&page={page}'
    
    # Make the API request and retrieve the movie data
    try:
        response = requests.get(url)
        movies = response.json()['results']
        
        # Convert the movie data to a DataFrame and append it to the movies_df
        movies_df = pd.concat([movies_df, pd.DataFrame(movies)], ignore_index=True)
        
    except:
        # If there is an error with the API request or response, print an error message
        print(f'Error on page {page}')
```
